```python
import matplotlib.pyplot as plt
import numpy as np

# Example 1: Line Plot
x_values = np.linspace(0, 10, 100)
y_values = np.sin(x_values)

plt.figure(figsize=(8, 5))
plt.plot(x_values, y_values, label='sin(x)')
plt.title('Line Plot')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.legend()
plt.grid(True)
plt.show()

# Example 2: Scatter Plot
np.random.seed(42)
x_random = np.random.rand(50)
y_random = np.random.rand(50)

plt.figure(figsize=(8, 5))
plt.scatter(x_random, y_random, color='blue', marker='o', label='Random Points')
plt.title('Scatter Plot')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.legend()
plt.grid(True)
plt.show()

# Example 3: Bar Chart
categories = ['Category A', 'Category B', 'Category C']
values = [25, 50, 75]

plt.figure(figsize=(8, 5))
plt.bar(categories, values, color='green', alpha=0.7)
plt.title('Bar Chart')
plt.xlabel('Categories')
plt.ylabel('Values')
plt.grid(axis='y')
plt.show()
```
